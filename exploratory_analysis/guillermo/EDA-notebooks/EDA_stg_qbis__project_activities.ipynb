# Exploratory Data Analysis - Employee Dimension Data
This notebook performs EDA on the `dim_employees_anon.csv` file from the Data Edge Internal Data folder.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
# Set Data Directory Path
data_dir = r"c:\Users\guine\Documents\BC#4\Hyper_python\BI project\Data Edge Internal Data"
file_path = os.path.join(data_dir, "stg_qbis__project_activities.csv")

# Verify file exists
if os.path.exists(file_path):
    print(f"✓ File found: {file_path}")
    print(f"File size: {os.path.getsize(file_path)} bytes")
else:
    print(f"✗ File not found: {file_path}")
```

```python
# Load Employee Data
try:
    df_project_activities = pd.read_csv(file_path)
    print("✓ Data loaded successfully!")
    print(f"Dataset shape: {df_project_activities.shape}")
except Exception as e:
    print(f"✗ Error loading data: {e}")
```

## Dataset Overview

```python
# Basic Information
print("=== BASIC INFORMATION ===")
print(f"Shape: {df_project_activities.shape}")
print(f"Columns: {list(df_project_activities.columns)}")
print(f"Data types:\n{df_project_activities.dtypes}")
print(f"\nMemory usage: {df_project_activities.memory_usage(deep=True).sum() / 1024:.2f} KB")
```

```python
# First Few Rows
print("=== FIRST 5 ROWS ===")
display(df_project_activities.head())

print("\n=== LAST 5 ROWS ===")
display(df_project_activities.tail())
```

## Data Quality Assessment

```python
# Missing Values Analysis
print("=== MISSING VALUES ===")
missing_data = df_project_activities.isnull().sum()
missing_percent = (missing_data / len(df_project_activities)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage': missing_percent
}).sort_values('Missing Count', ascending=False)

print(missing_df[missing_df['Missing Count'] > 0])

# Duplicates
print(f"\n=== DUPLICATES ===")
print(f"Total duplicated rows: {df_project_activities.duplicated().sum()}")
print(f"Percentage of duplicates: {(df_project_activities.duplicated().sum() / len(df_project_activities)) * 100:.2f}%")
```

## Statistical Summary

```python
# Statistical Summary for Numerical Columns
print("=== NUMERICAL COLUMNS SUMMARY ===")
numerical_cols = df_project_activities.select_dtypes(include=[np.number]).columns
if len(numerical_cols) > 0:
    display(df_project_activities[numerical_cols].describe())
else:
    print("No numerical columns found")

print("\n=== CATEGORICAL COLUMNS SUMMARY ===")
categorical_cols = df_project_activities.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  Unique values: {df_project_activities[col].nunique()}")
    print(f"  Top 5 values:\n{df_project_activities[col].value_counts().head()}")
```

```python
# Analyze processed_at date column
print("=== PROCESSED_AT DATE ANALYSIS ===")

# Check the current data type and sample values
print(f"Current data type: {df_project_activities['processed_at'].dtype}")
print(f"Sample values:")
print(df_project_activities['processed_at'].head(10))

# Convert processed_at to datetime
try:
    df_project_activities['processed_at_datetime'] = pd.to_datetime(df_project_activities['processed_at'])
    print(f"\n✓ Successfully converted to datetime")
    
    # Extract just the date part (without time)
    df_project_activities['processed_date'] = df_project_activities['processed_at_datetime'].dt.date
    
    # Find min and max dates
    min_datetime = df_project_activities['processed_at_datetime'].min()
    max_datetime = df_project_activities['processed_at_datetime'].max()
    min_date = df_project_activities['processed_date'].min()
    max_date = df_project_activities['processed_date'].max()
    
    print(f"\n=== DATE RANGE ANALYSIS ===")
    print(f"Minimum datetime: {min_datetime}")
    print(f"Maximum datetime: {max_datetime}")
    print(f"Minimum date: {min_date}")
    print(f"Maximum date: {max_date}")
    
    # Calculate the date range
    date_range = max_datetime - min_datetime
    print(f"Total date range: {date_range.days} days")
    
    # Count records per date
    date_counts = df_project_activities['processed_date'].value_counts().sort_index()
    print(f"\n=== RECORDS PER DATE ===")
    print(f"Total unique dates: {len(date_counts)}")
    print(f"Records per date (first 10):")
    display(date_counts.head(10).to_frame('Record Count'))
    
    # Show date distribution
    print(f"\nDate distribution summary:")
    print(f"  Most records in a single date: {date_counts.max()}")
    print(f"  Least records in a single date: {date_counts.min()}")
    print(f"  Average records per date: {date_counts.mean():.2f}")
    
    # Create a simple visualization
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    date_counts.plot(kind='line', figsize=(12, 6))
    plt.title('Records Over Time')
    plt.xlabel('Date')
    plt.ylabel('Number of Records')
    plt.xticks(rotation=45)
    
    plt.subplot(1, 2, 2)
    date_counts.hist(bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title('Distribution of Daily Record Counts')
    plt.xlabel('Number of Records per Day')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"✗ Error converting to datetime: {e}")
    print("Sample of problematic values:")
    print(df_project_activities['processed_at'].head(10))
```
