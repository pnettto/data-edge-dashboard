```python
# Analysis of Won vs. Lost Deals by Owner
print("=== DEAL WON/LOST RATIO BY OWNER ===")

# Check for required columns
if 'owner_id' in df_deals_fct.columns and 'deal_stage' in df_deals_fct.columns:
    print("✓ Required columns 'owner_id' and 'deal_stage' found.")
    
    # 1. Filter for only 'closedwon' and 'closedlost' deals
    relevant_deals = df_deals_fct[df_deals_fct['deal_stage'].isin(['closedwon', 'closedlost'])]
    
    if not relevant_deals.empty:
        # 2. Group by owner_id and count the occurrences of each deal stage
        # .unstack() pivots the deal stages into columns
        # fill_value=0 handles owners who might only have won or lost deals, but not both
        deal_counts = relevant_deals.groupby('owner_id')['deal_stage'].value_counts().unstack(fill_value=0)
        
        # Ensure both columns exist after unstacking, even if one had no deals
        if 'closedwon' not in deal_counts.columns:
            deal_counts['closedwon'] = 0
        if 'closedlost' not in deal_counts.columns:
            deal_counts['closedlost'] = 0
            
        # 3. Calculate the Won-to-Lost ratio
        # We replace 0 in 'closedlost' with NaN to handle division by zero.
        # The result will be 'inf' (infinity) for owners with wins but no losses.
        deal_counts['Won_to_Lost_Ratio'] = deal_counts['closedwon'] / deal_counts['closedlost'].replace(0, np.nan)
        
        # 4. Create the final table with the desired columns
        ratio_table = deal_counts[['Won_to_Lost_Ratio']].reset_index()
        
        # Sort the table for better readability, with highest ratios first
        ratio_table_sorted = ratio_table.sort_values(by='Won_to_Lost_Ratio', ascending=False).reset_index(drop=True)
        
        print("\n--- Ratio of Closed-Won to Closed-Lost Deals ---")
        print("(Note: 'inf' means the owner had wins but zero losses)")
        display(ratio_table_sorted)
        
    else:
        print("❌ No deals with 'closedwon' or 'closedlost' stages found in the data.")

else:
    print("❌ Missing 'owner_id' or 'deal_stage' column.")
```

```python
# Analysis of Won vs. Lost Deals by Owner (Corrected Win Ratio)
print("=== DEAL WIN RATIO BY OWNER ===")

# Check for required columns
if 'owner_id' in df_deals_fct.columns and 'deal_stage' in df_deals_fct.columns:
    print("✓ Required columns 'owner_id' and 'deal_stage' found.")
    
    # 1. Filter for only 'closedwon' and 'closedlost' deals
    relevant_deals = df_deals_fct[df_deals_fct['deal_stage'].isin(['closedwon', 'closedlost'])]
    
    if not relevant_deals.empty:
        # 2. Group by owner_id and count the occurrences of each deal stage
        deal_counts = relevant_deals.groupby('owner_id')['deal_stage'].value_counts().unstack(fill_value=0)
        
        # Ensure both 'closedwon' and 'closedlost' columns exist
        if 'closedwon' not in deal_counts.columns:
            deal_counts['closedwon'] = 0
        if 'closedlost' not in deal_counts.columns:
            deal_counts['closedlost'] = 0
            
        # 3. Calculate the total number of closed deals for each owner
        deal_counts['total_closed'] = deal_counts['closedwon'] + deal_counts['closedlost']
        
        # 4. CORRECTED: Calculate the Win Ratio (Wins / Total Closed Deals)
        # This will give a value between 0.0 (0% wins) and 1.0 (100% wins)
        deal_counts['Win_Ratio'] = deal_counts['closedwon'] / deal_counts['total_closed']
        
        # Replace NaN with 0 if an owner has 0 total deals (edge case)
        deal_counts['Win_Ratio'].fillna(0, inplace=True)
        
        # 5. Create the final table with more context
        ratio_table = deal_counts[['closedwon', 'closedlost', 'total_closed', 'Win_Ratio']].reset_index()
        
        # Sort the table by the highest win ratio
        ratio_table_sorted = ratio_table.sort_values(by='Win_Ratio', ascending=False).reset_index(drop=True)
        
        print("\n--- Win Ratio by Owner ---")
        print("(Win Ratio = Deals Won / Total Closed Deals)")
        
        # Display the table with the ratio formatted as a percentage
        display(ratio_table_sorted)
        
    else:
        print("❌ No deals with 'closedwon' or 'closedlost' stages found in the data.")

else:
    print("❌ Missing 'owner_id' or 'deal_stage' column.")
```

```python

# Check for required columns
required_cols = ['create_date', 'close_date', 'deal_stage', 'deal_amount']
if all(col in df_deals_fct.columns for col in required_cols):
    
     # 1. Prepare the data for survival analysis
    survival_df = df_deals_fct[required_cols].copy()
    
    # Convert date columns to datetime, coercing errors
    survival_df['create_date_dt'] = pd.to_datetime(survival_df['create_date'], errors='coerce')
    survival_df['close_date_dt'] = pd.to_datetime(survival_df['close_date'], errors='coerce')
    
    # Drop rows with invalid or missing dates or deal_amount
    survival_df.dropna(subset=['create_date_dt', 'close_date_dt', 'deal_amount'], inplace=True)
    
    # Calculate the duration of the deal in days
    survival_df['duration'] = (survival_df['close_date_dt'] - survival_df['create_date_dt']).dt.days

    # The "event" is a deal being won. 
    survival_df['event_observed'] = (survival_df['deal_stage'] == 'closedwon').astype(int)
    
    # Filter out deals with non-positive duration
    survival_df = survival_df[survival_df['duration'] > 0]
    
    # --- NEW: FEATURE ENGINEERING - Create Deal Amount Buckets ---
    # Use qcut to create 4 buckets (quartiles) with roughly equal numbers of deals
    try:
        survival_df['deal_amount_bucket'] = pd.qcut(
            survival_df['deal_amount'], 
            q=4, 
            labels=['Small', 'Medium', 'Large', 'X-Large'],
            duplicates='drop' # Handles cases where bin edges are not unique
        )
        print("\n✓ Successfully created deal amount buckets.")
        print("Bucket distribution:")
        print(survival_df['deal_amount_bucket'].value_counts())
    except Exception as e:
        print(f"✗ Could not create buckets: {e}")
        survival_df['deal_amount_bucket'] = 'default' # Fallback
    
    print(f"\nPrepared {len(survival_df)} deals for analysis.")
    
     # 2. Kaplan-Meier Survival Curve (Overall)

    print("\n--- Kaplan-Meier Survival Curve ---")

    print("This curve shows the probability of a deal *not yet being won* over time.")

   

    kmf = KaplanMeierFitter()

    kmf.fit(durations=survival_df['duration'], event_observed=survival_df['event_observed'])

   

    plt.figure(figsize=(12, 7))

    kmf.plot(label='Probability of Deal Remaining "Not Won"')

   

    plt.title('Survival Function: Time to Win a Deal', fontsize=16, fontweight='bold')

    plt.xlabel('Deal Duration (Days)', fontsize=12)

    plt.ylabel('Probability of Survival (Not Won)', fontsize=12)

    plt.grid(True, which='major', linestyle='--', alpha=0.6)

    plt.legend()

    plt.show()

   

    # Print median survival time

    median_survival_time = kmf.median_survival_time_

    print(f"Median time to win a deal: {median_survival_time:.0f} days")

    print("(This means 50% of deals that are eventually won will be won by this day.)")


    # 3. Cox Proportional Hazards Model (Using deal_amount_bucket)
    print("\n--- Cox Proportional Hazards Model ---")
    print("This model identifies if the deal amount bucket influences the speed of winning a deal.")
    
    # Select features for the model
    cox_df = survival_df[['duration', 'event_observed', 'deal_amount_bucket']].copy()
    
    # Convert the categorical bucket feature into dummy variables
    cox_df_dummies = pd.get_dummies(cox_df, columns=['deal_amount_bucket'], drop_first=True)

    if not cox_df_dummies.empty and 'deal_amount_bucket_Medium' in cox_df_dummies.columns:
        # The lifelines library needs to be imported if not already done
        try:
            from lifelines import CoxPHFitter
        except ImportError:
            print("✗ 'lifelines' library not found. Please run: !pip install lifelines")
            exit()

        cph = CoxPHFitter(penalizer=0.1)
        cph.fit(cox_df_dummies, duration_col='duration', event_col='event_observed')
        
        print("\nModel Summary:")
        cph.print_summary()
        
        # Plot the coefficients
        plt.figure(figsize=(10, 6))
        cph.plot()
        plt.title('Cox Model: Impact of Deal Amount Bucket on Winning a Deal', fontsize=16, fontweight='bold')
        plt.show()
    else:
        print("❌ Could not generate a valid model. Check bucket creation.")

```

```python
# Cohort Analysis: Tracking Win Rate Over Time
print("=== COHORT ANALYSIS: CUMULATIVE WIN RATE BY CREATION QUARTER ===")

# Check for required columns
required_cols = ['deal_id', 'create_date', 'close_date', 'deal_stage']
if all(col in df_deals_fct.columns for col in required_cols):
    print("✓ Required columns found.")
    
    # 1. Prepare the data
    cohort_df = df_deals_fct[required_cols].copy()
    
    # Convert date columns to datetime, coercing errors
    cohort_df['create_date_dt'] = pd.to_datetime(cohort_df['create_date'], errors='coerce')
    cohort_df['close_date_dt'] = pd.to_datetime(cohort_df['close_date'], errors='coerce')
    
    # Drop rows with invalid or missing dates
    cohort_df.dropna(subset=['create_date_dt', 'close_date_dt'], inplace=True)
    
    # Filter for only closed deals, as they are the only ones with a final outcome
    cohort_df = cohort_df[cohort_df['deal_stage'].isin(['closedwon', 'closedlost'])]
    
    # 2. Assign Cohorts and Timestamps
    # The cohort is the quarter the deal was created in
    cohort_df['cohort_quarter'] = cohort_df['create_date_dt'].dt.to_period('Q')
    
    # The quarter the deal was closed (won or lost)
    cohort_df['close_quarter'] = cohort_df['close_date_dt'].dt.to_period('Q')
    
    # Calculate the 'age' of a deal in quarters at the time of closing
    cohort_df['quarters_to_close'] = (cohort_df['close_quarter'] - cohort_df['cohort_quarter']).apply(lambda x: x.n)
    
    # Filter out any deals that closed before they were created
    cohort_df = cohort_df[cohort_df['quarters_to_close'] >= 0]
    
    print(f"\nAnalyzing {len(cohort_df)} closed deals across {cohort_df['cohort_quarter'].nunique()} cohorts.")

    # 3. Calculate Cohort Metrics
    # Get the total number of deals in each cohort
    cohort_sizes = cohort_df.groupby('cohort_quarter')['deal_id'].nunique()
    
    # Filter for only 'closedwon' deals to calculate wins
    wins_df = cohort_df[cohort_df['deal_stage'] == 'closedwon']
    
    # Group by cohort and closing quarter to count wins over time
    cohort_wins = wins_df.groupby(['cohort_quarter', 'quarters_to_close'])['deal_id'].nunique().reset_index()
    
    # Calculate the cumulative number of wins for each cohort over time
    cohort_wins['cumulative_wins'] = cohort_wins.groupby('cohort_quarter')['deal_id'].cumsum()
    
    # Merge cohort sizes back to calculate the win rate
    cohort_analysis = pd.merge(cohort_wins, cohort_sizes.rename('total_deals'), on='cohort_quarter')
    
    # Calculate the cumulative win rate
    cohort_analysis['cumulative_win_rate'] = cohort_analysis['cumulative_wins'] / cohort_analysis['total_deals']
    
    # 4. Create the Cohort Heatmap
    # Pivot the table to create the matrix for the heatmap
    cohort_pivot = cohort_analysis.pivot_table(
        index='cohort_quarter',
        columns='quarters_to_close',
        values='cumulative_win_rate'
    )
    
    # Get the cohort size for labeling the heatmap
    cohort_sizes_formatted = cohort_sizes.map('{:,.0f}'.format)

    plt.figure(figsize=(18, 10))
    
    # Generate the heatmap
    ax = sns.heatmap(
        cohort_pivot,
        annot=True,
        fmt='.1%', # Format annotations as percentages
        cmap='viridis',
        cbar_kws={'label': 'Cumulative Win Rate'}
    )
    
    # Set titles and labels
    ax.set_title('Cumulative Win Rate by Deal Creation Cohort', fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel('Quarters Since Deal Creation', fontsize=12)
    ax.set_ylabel('Deal Creation Cohort', fontsize=12)
    
    # Add cohort sizes to the y-axis labels
    ax.set_yticklabels([f'{label.get_text()} (n={cohort_sizes_formatted.get(pd.Period(label.get_text(), "Q"), 0)})' for label in ax.get_yticklabels()])
    
    plt.tight_layout()
    plt.show()
    
    print("\n--- How to Read This Heatmap ---")
    print("Each row represents a cohort of deals created in that quarter.")
    print("Each column represents the number of quarters that have passed since creation.")
    print("The color and percentage show the cumulative win rate for that cohort at that point in time.")
    print("For example, you can track the 2023Q1 cohort row to see how its win rate increased over subsequent quarters.")

else:
    print(f"❌ Missing one or more required columns: {required_cols}")
```
